Welcome to the notebook tutorial! In this comprehensive guide, we will walk you through the step-by-step instructions for setting up a complete data science environment using R, Rstudio, and Jupyter Notebook.

<font size="4">**Installing R and Rstudio**</font>

*   **R**: To download R, navigate to the Comprehensive R Archive Network, or CRAN, at https://cloud.r-project.org. CRAN offers tailored versions of R for various operating systems including Windows, Mac, and Linux, ensuring broad accessibility. For the purposes of this course, we recommend using R version 4.2.0 or later.
*   **RStudio**: This is an Integrated Development Environment (IDE) for R. It provides a user-friendly interface and powerful tools for using R. It can be downloaded from https://posit.co/download/rstudio-desktop/. RStudio is basically the way you will interact with R!

<font size="4">**Rstudio Environment**</font>

When you open R-Studio you will see a window with four panes. Here are
the default positions:


*   **Top-Left**: The source pane where you can open/edit code files and save for future use.
*   **Bottom-Left**: The console where you can type in the codes and where R prints the numeric outputs.
*   **Top-Right**: The “Environment” pane where the information of any object (scalars, vectors or data-sets) created in any session show up.
*   **Bottom-Right**: The “plots” pane where any graphical output will show up.
